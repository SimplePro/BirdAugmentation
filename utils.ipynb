```python
# 1. CUB200 데이터셋부터 채우기

from config import class_imgid_list, class_name_list, imgid_filename_dict, imgid_bbox_dict
from tqdm import tqdm
from PIL import Image

class_dict = {
    "crow": [29, 30, 107, 108], 
    "cuckoo": [31, 33],
    "sparrow": [113, 123, 127, 129, 130],
    "swallow": [136, 137, 138],
    "woodpecker": [188, 189, 190, 191, 192]
}

for i, class_name in tqdm(enumerate(list(class_dict.keys()))):
    for c in class_dict[class_name]:
        
        for imgid in class_imgid_list[c]:
            
            file_name = imgid_filename_dict[imgid]

            bird_path = os.path.join("CUB200", "images", class_name_list[c], file_name)
            bird_img = Image.open(bird_path)
            W, H = bird_img.size
            bird_img = bird_img.resize((256, 256)).convert("RGB")
            bird_images.append(bird_img)

            mask_path = os.path.join("CUB200", "segmentations", class_name_list[c], file_name.replace("jpg", "png"))
            mask_img = Image.open(mask_path).resize((256, 256)).convert("L")
            mask_images.append(mask_img)

            h_scailing, w_scailing = 256 / H, 256 / W
            xmin, ymin, w, h = imgid_bbox_dict[imgid]
            xmin, ymin, w, h = xmin * w_scailing, ymin * h_scailing, w * w_scailing, h * h_scailing
            xmin, ymin, w, h = xmin / 256, ymin / 256, w / 256, h / 256
            x, y = (xmin + w/2), (ymin + h/2)

            bboxes_list.append("\n".join([f"{i} {x} {y} {w} {h}"]) + "\n")
```

```python
# 2. crawling dataset 채우기

from glob import glob

root_dir = "/home/kdhsimplepro/kdhsimplepro/AI/BirdAugmentation"

bird_images_paths = []
mask_images_paths = []
bboxes_paths = []

for class_name in class_dict.keys():
    for img_path in glob(os.path.join(root_dir, "crawling_dataset", class_name, "obj_train_data", "*.jpg")):
        id_ = img_path.split('/')[-1].split(".")[0]
        
        bird_images_paths.append(img_path)
        mask_images_paths.append(os.path.join(root_dir, "BirdSegmentation", "mask_crawling_images", class_name, f"{id_}.jpg"))
        bboxes_paths.append(os.path.join(root_dir, "crawling_dataset", class_name, "obj_train_data", f"{id_}.txt"))


for bird_image_path, mask_image_path, bboxes_path in zip(bird_images_paths, mask_images_paths, bboxes_paths):
    bird_image = Image.open(bird_image_path)
    mask_image = Image.open(mask_image_path)
    with open(bboxes_path, "r") as f:
        lines = f.readlines()
        bboxes = [list(map(float, line.replace("\n", "").split())) for line in lines]
    
    bird_image = bird_image.resize((256, 256)).convert("RGB")
    mask_image = mask_image.resize((256, 256)).convert("L")

    for i in range(len(bboxes)):
        c = int(bboxes[i][0])
        x, y, w, h = bboxes[i][1:]
        bboxes[i] = f"{c} {x} {y} {w} {h}"

    bboxes = "\n".join(bboxes) + "\n"

    bird_images.append(bird_image)
    mask_images.append(mask_image)
    bboxes_list.append(bboxes)
```

```python
from PIL import ImageDraw
from copy import deepcopy

def draw_bbox(image, bboxes):
    img = deepcopy(image)
    W, H = img.size

    draw = ImageDraw.Draw(img)

    for x, y, w, h in bboxes:
        xmin = (x - (w/2)) * W
        ymin = (y - (h/2)) * H
        xmax = (x + (w/2)) * W
        ymax = (y + (h/2)) * H

        draw.rectangle((xmin, ymin, xmax, ymax), outline=(255, 0, 0), width=3)
    
    return img
```

```python
import albumentations as A
from random import uniform
import torch.nn.functional as F

class DataAugmentator:

    def __init__(
        self,
        bird_images,
        mask_images,
        bboxes_list,
        min_size_of_bird=8,
        max_size_of_bird=256,
    ):
    
        self.img_H, self.img_W = 256, 256

        self.bird_images = bird_images
        self.mask_images = mask_images
        self.bboxes_list = bboxes_list

        self.min_size_of_bird = min_size_of_bird
        self.max_size_of_bird = max_size_of_bird

        self.croped_bird_images = []
        self.croped_mask_images = []

        self.crop_bird_and_mask_images()

        self.albumentations_dict = {
            "crop_and_pad": A.CropAndPad(percent=(-0.2,0.4), p=1),
            "rotate": A.SafeRotate(p=0.5),
            "vertical_flip": A.VerticalFlip(p=0.5),
            "horizontal_flip": A.HorizontalFlip(p=0.5),
            "random_brightness_contrast": A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=0.8),
            "blur": A.AdvancedBlur(p=0.3),
            "fog": A.RandomFog(p=0.3),
        }

    
    def crop_bird_and_mask_images(
        self
    ):
        
        for bird_img, mask_img, bboxes in tqdm(zip(self.bird_images, self.mask_images, self.bboxes_list)):
            
            for c, x, y, w, h in bboxes:
                xmin, ymin, xmax, ymax = (x - w/2), (y - h/2), (x + w/2), (y + h/2)
                xmin, ymin, xmax, ymax = xmin * self.img_W, ymin * self.img_H, xmax * self.img_W, ymax * self.img_H

                self.croped_bird_images.append(
                    bird_img[:, ymin:ymax, xmin:xmax]
                )

                self.croped_mask_images.append(
                    mask_img[:, ymin:ymax, xmin:xmax]
                )


    def do_augmentation_img(
        self,
        bird_img,
        mask_img,
        bboxes,
        compose_list=["rotate", "vertical_flip", "horizontal_flip", "random_brightness_contrast", "blur", "fog"]
    ):
        
        transformation = A.Compose([
            self.albumentations_dict[key]
            for key in compose_list
        ], bbox_params=A.BboxParams(format="yolo", label_fields=["class_labels"]))

        albumentation_result = transformation(
            image=bird_img.permute(1, 2, 0).numpy(),
            mask=mask_img.permute(1, 2, 0).numpy(),
            bboxes=[bbox[1:] for bbox in bboxes],
            class_labels=[bbox[0] for bbox in bboxes]
        )

        return (
            torch.from_numpy(albumentation_result["image"]),
            torch.from_numpy(albumentation_result["mask"]),
            [[bboxes[i][0]] + list(albumentation_result[i]) for i in range(len(bboxes))]
        )



    def random_resize(
        self,
        bird_img,
        mask_img,
    ):
        bird_H, bird_W = bird_img.size(1), bird_img.size(2)
        
        min_scale_factor = self.min_size_of_bird / min(bird_H, bird_W)
        max_scale_factor = self.max_size_of_bird / max(bird_H, bird_W)

        random_scale_factor = uniform(min_scale_factor, max_scale_factor)

        resized_bird_img = F.interpolate(bird_img, scale_factor=random_scale_factor)
        resized_mask_img = F.interpolate(mask_img, scale_factor=random_scale_factor)

        return resized_bird_img, resized_mask_img


    def random_position(
        self,
        bird_img,
        mask_img,
        background_img
    ):
        pass

    def paste_img(
        self
    ):
        pass
```
