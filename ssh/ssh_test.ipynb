```python
from glob import glob

train_img_list = glob("/home/kdhsimplepro/kdhsimplepro/AI/BirdAugmentation/ssh/BirdAugmentation/dataset/train/images/*.jpg")
valid_img_list = glob("/home/kdhsimplepro/kdhsimplepro/AI/BirdAugmentation/ssh/BirdAugmentation/dataset/test/images/*.jpg")
print(len(train_img_list), len(valid_img_list))
```

```python
from random import shuffle

shuffle(train_img_list)
shuffle(valid_img_list)

print(train_img_list[:10])
print(valid_img_list[:10])
```

```python
with open("./BirdAugmentation/dataset/train.txt", "w") as f:
    f.write("\n".join(train_img_list) + "\n")

with open("./BirdAugmentation/dataset/val.txt", "w") as f:
    f.write("\n".join(valid_img_list) + "\n")
```

```python
# 데이터셋의 yaml 파일 확인

import yaml

with open("./BirdAugmentation/dataset/data.yaml", "r") as f:
    data = yaml.safe_load(f)

data["train"] = "/home/kdhsimplepro/kdhsimplepro/AI/BirdAugmentation/ssh/BirdAugmentation/dataset/train.txt"
data["val"] = "/home/kdhsimplepro/kdhsimplepro/AI/BirdAugmentation/ssh/BirdAugmentation/dataset/val.txt"

with open("./BirdAugmentation/dataset/data.yaml", "w") as f:
    yaml.dump(data, f)

print(data)
```
